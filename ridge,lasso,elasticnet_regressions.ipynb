```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('mpg')
```

```python
df
```

```python
df.drop('name',axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
df['horsepower'].median()
```

```python
df['horsepower']=df['horsepower'].fillna(df['horsepower'].median())
```

```python
df
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.origin.unique()
```

```python
df['origin'].value_counts()
```

```python
df['origin']=df['origin'].map({'usa':1,'japan':2,'europe':3})
```

```python
df
```

```python
x=df.drop('mpg',axis=1)
y=df['mpg']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
for i,col_name in enumerate(x_train.columns):
  print(f'the coefficient for {col_name}is {model.coef_[i]}')
```

```python
from sklearn.metrics import r2_score
y_pred=model.predict(x_test)
r2_score(y_test,y_pred)
```

```python
# ridge_regression:
from sklearn.linear_model import Ridge
ridge_regression_model=Ridge(alpha=0.1)
ridge_regression_model.fit(x_train,y_train)
for i,col_name in enumerate(x_train.columns):
  print(f'the coefficient for {col_name}is {ridge_regression_model.coef_[i]}')
```

```python
from sklearn.metrics import r2_score
y_pred=ridge_regression_model.predict(x_test)
r2_score(y_test,y_pred)
```

```python
# lasso_regression:
from sklearn.linear_model import Lasso
lasso_regression_model=Lasso(alpha=0.1)
lasso_regression_model.fit(x_train,y_train)
for i,col_name in enumerate(x_train.columns):
  print(f'the coefficient for {col_name}is {lasso_regression_model.coef_[i]}')
```

```python
from sklearn.metrics import r2_score
y_pred=lasso_regression_model.predict(x_test)
r2_score(y_test,y_pred)
```

```python
# ElasticNet_regression:
from sklearn.linear_model import ElasticNet
ElasticNet_regression_model=ElasticNet(alpha=1,l1_ratio=0.5)
ElasticNet_regression_model.fit(x_train,y_train)
for i,col_name in enumerate(x_train.columns):
  print(f'the coefficient for {col_name}is {ElasticNet_regression_model.coef_[i]}')
```

```python
from sklearn.metrics import r2_score
y_pred=ElasticNet_regression_model.predict(x_test)
r2_score(y_test,y_pred)
```
